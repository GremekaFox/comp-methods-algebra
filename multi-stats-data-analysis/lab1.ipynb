
2. Вычислить числовые характеристики выборки: среднее, стандартное отклонение, дисперсию,
медиану, первую и третью квартиль, коэффициент асимметрии, эксцесса, минимальное и
максимальное значение выборки. Какие можно сделать выводы?

```python
def get_stats(df):
    stats = df.describe().T
    stats["median"] = df.numbers.median()
    stats["dispersion"] = stats.loc[:, "std"][0] ** 2
    stats["skew"] = df.numbers.skew() # асимметрия
    stats["kurtosis"] = df.numbers.kurtosis() # эксцесс
    return stats

stats = get_stats(df)
stats
```

## 3. Проверкаa правила 3σ.

```python
# stats["mean"] - среднее
# stats["std"] - std

sigma_3_mean = df[(df.numbers < stats["mean"][0] + 3 * stats["std"][0]) & (df.numbers > stats["mean"][0] - 3 * stats["std"][0])]
sigma_3_median = df[(df.numbers < stats["median"][0] + 3 * stats["std"][0]) & (df.numbers > stats["median"][0] - 3 * stats["std"][0])]
```
